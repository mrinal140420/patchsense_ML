```python
import pandas as pd
from pathlib import Path
import os
import numpy as np

PROJECT_ROOT = Path.cwd()
while not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

df = pd.read_parquet(PROJECT_ROOT / "data" / "processed" / "nvd_full.parquet")
df.shape
```

```python
df["proxy_relevant"] = df["has_exploit_ref"].astype(int)

df["proxy_relevant"].mean()
```

```python
baseline_df = df[
    df["cvss"].notna()
].sort_values(
    by="cvss",
    ascending=False
).reset_index(drop=True)
```

```python
def precision_at_k(y_true, scores, k=1000):
    idx = np.argsort(scores)[::-1][:k]
    return y_true.iloc[idx].mean()
```

```python
precision_1000 = precision_at_k(
    baseline_df["proxy_relevant"],
    baseline_df["cvss"],
    k=1000
)

precision_1000
```
